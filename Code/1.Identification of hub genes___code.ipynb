## GEO33335

```python
# import modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
#workpath
os.chdir("./GEO_train/")
#GEO33335 soft file
GEO33335_info=pd.read_csv("./RAW/GSE33335_family.soft",sep="\t",skiprows=165,index_col=0)
```

```python
GEO33335_info.columns
```

```python
GEO33335_info=GEO33335_info.loc[:,["gene_assignment"]]
GEO33335_info=GEO33335_info.gene_assignment.replace("---", None)
GEO33335_info=GEO33335_info.dropna()
```

```python
#get gene symbol
def fun1(x):
    if "//" in x:
        return x.split("//")[1].strip()
GEO33335_info=GEO33335_info.apply(fun1)
GEO33335_info
```

```python
GEO33335_info_Matrix=pd.DataFrame({"gene_name":list(GEO33335_info)},index=list(GEO33335_info.index))
GEO33335_info_Matrix
```

```python
GEO33335_info_Matrix.to_csv("./result/GEO33335_info_Matrix.csv")
```

```python
#get samples classification
def readfile(matrix_file):
    with open(matrix_file,"r") as f:
       
        for line in f:
            
            if line.startswith("!Sample_description"):
                gene_class=line.rstrip().split("\t")
                       
    return gene_class       
GSE33335_sample_class=readfile("./RAW/GSE33335_series_matrix.txt")
GSE33335_sample_label=[]
GSE33335_sample_class=GSE33335_sample_class[1:]
GSE33335_normal=0
GSE33335_tumor=0
for i in range(len(GSE33335_sample_class)):
    if GSE33335_sample_class[i][1]=="n":
        GSE33335_sample_label.append(0)
        GSE33335_normal+=1
    elif GSE33335_sample_class[i][1]=="t":
        GSE33335_sample_label.append(1)
        GSE33335_tumor+=1
print(GSE33335_sample_label,GSE33335_normal,GSE33335_tumor)
```

```python
#import GEO33335gene Matrix
GEO33335_gene_Matrix=pd.read_csv("./RAW/GSE33335_series_matrix.txt",comment="!",sep="\t")
#import GEO33335gene Ann file
GEO33335_info_Matrix=pd.read_csv("./result/GEO33335_info_Matrix.csv",index_col=0)
GSE33335_ALL_gene_Matrix=GSE33335_gene_Matrix.join(GEO33335_info_Matrix, on='ID_REF')
#Pretreatment
GSE33335_ALL_gene_Matrix=GSE33335_ALL_gene_Matrix.dropna()
GSE33335_ALL_gene_Matrix.gene_name.value_counts()
GSE33335_ALL_gene_Matrix=GSE33335_ALL_gene_Matrix.groupby(by=GSE33335_ALL_gene_Matrix.gene_name).mean()
GSE33335_ALL_gene_Matrix.index.value_counts()
GSE33335_ALL_gene_Matrix=GSE33335_ALL_gene_Matrix.drop("ID_REF",axis=1)
GSE33335_ALL_gene_Matrix=GSE33335_ALL_gene_Matrix.T
#Build training set
GSE33335_ALL_gene_Matrix["label"]=GSE33335_sample_label
GSE33335_ALL_gene_Matrix
#save file 
GSE33335_ALL_gene_Matrix.to_csv("./result/GSE33335_ALL_gene_Matrix.csv")
```

## GSE79973

```python
GSE79973_info=pd.read_csv("./RAW/GSE79973_family.soft",sep='\t',skiprows=165)
GSE79973_info.columns
```

```python
GSE79973_info=GSE79973_info.loc[:,["ID","Gene Symbol"]]
GSE79973_info=GSE79973_info.dropna()

def get_genename(d):
    list_ID=[]
    list_gene_name=[]
    for i in range(len(d)):
        if "///" in d.iloc[i,1]:
            str1=d.iloc[i,1].split("///")

            for j in range(len(str1)):
                list_ID.append(d.iloc[i,0])
                list_gene_name.append(str1[j])

        else:
            list_ID.append(d.iloc[i,0])
            list_gene_name.append(d.iloc[i,1])
    f=pd.DataFrame({"gene_name":list_gene_name},index=list_ID)
    return f
GSE79973_info_Matrix=get_genename(GSE79973_info)
GSE79973_info_Matrix
GSE79973_info_Matrix.to_csv("./result/GSE79973_info_Matrix.csv")
```

```python
def read_Matrix(Matrix_file):
    
    with open(Matrix_file,"r",encoding='gb18030', errors='ignore') as f:
       
        for line in f:
            
            if line.startswith("!Sample_title"):
                gene_class=line.rstrip().split("\t")
                       
    return gene_class  
              
GSE79973_sample_class=read_Matrix("./RAW/GSE79973_series_matrix.txt")
print(GSE79973_sample_class)
GSE79973_sample_label=[]
GSE79973_sample_class=GSE79973_sample_class[1:]
GSE79973_normal=0
GSE79973_tumor=0
for i in range(len(GSE79973_sample_class)):
    name_class=GSE79973_sample_class[i].split("_")[1][0]
    
    if name_class=="n":
        GSE79973_sample_label.append(0)
        GSE79973_normal+=1
    elif name_class=="t":
        GSE79973_sample_label.append(1)
        GSE79973_tumor+=1
        
print(GSE79973_sample_label,GSE79973_normal,GSE79973_tumor)
```

```python
GSE79973_gene_Matrix=pd.read_csv("./RAW/GSE79973_series_matrix.txt",comment="!",sep="\t")
GSE79973_ALL_gene_Matrix=GSE79973_gene_Matrix.join(GSE79973_info_Matrix,on="ID_REF")
GSE79973_ALL_gene_Matrix=GSE79973_ALL_gene_Matrix.dropna()
GSE79973_ALL_gene_Matrix=GSE79973_ALL_gene_Matrix.groupby(by=GSE79973_ALL_gene_Matrix.gene_name).mean()
GSE79973_ALL_gene_Matrix=GSE79973_ALL_gene_Matrix.T
GSE79973_ALL_gene_Matrix["label"]=GSE79973_sample_label
GSE79973_ALL_gene_Matrix.to_csv("./result/GSE79973_ALL_gene_Matrix.csv")
```

```python
def read_class(file_info):
    with open(file_info,"r",encoding='gb18030', errors='ignore') as f:
        for line in f:
            
            if line.startswith("!Sample_title"):
                gene_class=line.rstrip().split("\t")
        return  gene_class 
GSE658019_sample_class=read_class("./RAW/GSE65801_series_matrix.txt")
print(GSE658019_sample_class)
GSE65801_sample_class=GSE65801_sample_class[1:]
GSE65801_sample_label=[]
normal=0
tumor=0
for class_name in GSE65801_sample_class:
    s=class_name.split("_")[1]
    if s[0]=="n":
        GSE65801_sample_label.append(0)
        normal+=1
    else:
        GSE65801_sample_label.append(1)
        tumor+=1
print(GSE65801_sample_label,normal,tumor)
```

```python
GSE65801_gene_Matrix=pd.read_csv("./RAW/GSE65801_series_matrix.txt",comment="!",sep="\t")
GSE65801_info_Matrix=pd.read_csv("./result/GSE65801_info_Matrix.csv",index_col=0)
GSE65801_ALL_Matrix=GSE65801_gene_Matrix.join(GSE65801_info_Matrix,on="ID_REF")
GSE65801_ALL_Matrix=GSE65801_ALL_Matrix.dropna()
GSE65801_ALL_Matrix=GSE65801_ALL_Matrix.groupby(by="GENE_SYMBOL").mean()
GSE65801_test_Matrix=GSE65801_ALL_Matrix.T
GSE65801_test_Matrix["label"]=GSE65801_sample_label
GSE65801_test_Matrix
```
